## Problems

Now write and execute SQL queries to solve assignment problems

### Problem 1

##### Find the total number of crimes recorded in the CRIME table.


```python
stmt=ibm_db.exec_immediate(conn,'SELECT COUNT(ID) from CHICAGO_CRIME_DATA');
ibm_db.fetch_both(stmt)
```

### Problem 2

##### List community areas with per capita income less than 11000.


```python
stmt=ibm_db.exec_immediate(conn,'SELECT COMMUNITY_AREA_NAME from CENSUS_DATA \
where PER_CAPITA_INCOME <11000');
#ibm_db.fetch_assoc(stmt)
#Fetch the rest of the rows and print the ID and FNAME for those rows
while ibm_db.fetch_row(stmt) != False:
   print (" NAME:",  ibm_db.result(stmt, "COMMUNITY_AREA_NAME"))
```

### Problem 3

##### List all case numbers for crimes  involving minors?(children are not considered minors for the purposes of crime analysis)


```python
stmt = ibm_db.exec_immediate(conn, 'SELECT CASE_NUMBER from CHICAGO_CRIME_DATA \
            where DESCRIPTION LIKE \'%MINOR%\'');
while ibm_db.fetch_row(stmt) != False:
   print (ibm_db.result(stmt, "CASE_NUMBER"))                     
```

### Problem 4

##### List all kidnapping crimes involving a child?


```python
stmt=ibm_db.exec_immediate(conn, 'SELECT CASE_NUMBER from CHICAGO_CRIME_DATA \
            where PRIMARY_TYPE = \'KIDNAPPING\' and DESCRIPTION LIKE \'%CHILD%\'')
while ibm_db.fetch_row(stmt) != False:
    print (ibm_db.result(stmt, "CASE_NUMBER"))
```

### Problem 5

##### What kinds of crimes were recorded at schools?


```python
stmt=ibm_db.exec_immediate(conn, 'SELECT distinct(PRIMARY_TYPE) from CHICAGO_CRIME_DATA where LOCATION_DESCRIPTION like \'%SCHOOL%\'')
while ibm_db.fetch_row(stmt) != False:
    print (ibm_db.result(stmt, "PRIMARY_TYPE"))
```

### Problem 6

##### List the average safety score for all types of schools.


```python
stmt=ibm_db.exec_immediate(conn, 'SELECT AVG(SAFETY_SCORE) from CHICAGO_PUBLIC_SCHOOLS')
ibm_db.fetch_both(stmt)
```

```python
#Or for different type of schools each average
stmt=ibm_db.exec_immediate(conn, 'SELECT \"Elementary, Middle, or High School\",\
        AVG(SAFETY_SCORE) as AVG_SAFETY_SCORE \
        from CHICAGO_PUBLIC_SCHOOLS group by \"Elementary, Middle, or High School"')
while ibm_db.fetch_row(stmt) != False:
    print (ibm_db.result(stmt, "Elementary, Middle, or High School"),
           ibm_db.result(stmt,"AVG_SAFETY_SCORE"))
```

### Problem 7

##### List 5 community areas with highest % of households below poverty line


```python
stmt=ibm_db.exec_immediate(conn, 'SELECT COMMUNITY_AREA_NAME, PERCENT_HOUSEHOLDS_BELOW_POVERTY \
        from CENSUS_DATA order by PERCENT_HOUSEHOLDS_BELOW_POVERTY desc NULLS LAST LIMIT 5')
while ibm_db.fetch_row(stmt) != False:
    print (ibm_db.result(stmt, "COMMUNITY_AREA_NAME"),
           ibm_db.result(stmt,"PERCENT_HOUSEHOLDS_BELOW_POVERTY"))
```

### Problem 8

##### Which community area is most crime prone?


```python
stmt=ibm_db.exec_immediate(conn, 'SELECT COMMUNITY_AREA_NUMBER, count(*) AS CASE_AMOUNT \
    from CHICAGO_CRIME_DATA group by COMMUNITY_AREA_NUMBER \
    order by CASE_AMOUNT desc NULLS LAST LIMIT 1')
while ibm_db.fetch_row(stmt) != False:
    print (ibm_db.result(stmt, "COMMUNITY_AREA_NUMBER"),
           ibm_db.result(stmt,"CASE_AMOUNT"))
```

### Problem 9

##### Use a sub-query to find the name of the community area with highest hardship index


```python
stmt=ibm_db.exec_immediate(conn, 'SELECT COMMUNITY_AREA_NAME, HARDSHIP_INDEX \
        from CENSUS_DATA where HARDSHIP_INDEX = \
        (select MAX(HARDSHIP_INDEX) from CENSUS_DATA)')
while ibm_db.fetch_row(stmt) !=False:
    print(ibm_db.result(stmt, "COMMUNITY_AREA_NAME"),
          ibm_db.result(stmt,"HARDSHIP_INDEX"))
```

### Problem 10

##### Use a sub-query to determine the Community Area Name with most number of crimes?


```python
stmt=ibm_db.exec_immediate(conn, 
                           'CREATE VIEW NUM_CRIME AS \
                           SELECT COMMUNITY_AREA_NUMBER, count(*) \
                           as crime_amount from CHICAGO_CRIME_DATA \
                           group by COMMUNITY_AREA_NUMBER')
stmt=ibm_db.exec_immediate(conn,
                           'SELECT COMMUNITY_AREA_NAME from CENSUS_DATA\
                           where COMMUNITY_AREA_NUMBER = \
                           (select COMMUNITY_AREA_NUMBER from NUM_CRIME \
                           where crime_amount = (select max(crime_amount) \
                           from NUM_CRIME))')
while ibm_db.fetch_row(stmt) !=False:
    print(ibm_db.result(stmt, "COMMUNITY_AREA_NAME"))
```
